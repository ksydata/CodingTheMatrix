```python
# 1.1. 그림 1-2 벡터의 덧셈과 뺄셈
# 이동(걷기) 해석 & 평행사변형 법칙
# 벡터의 덧셈 : v만큼 걷고, 거기에서 w만큼 이어서 더 걷기
# 벡터의 뺄셈 : w끝점에서 v끝점까지의 거리

v = np.array([1, 2])
w = np.array([4, -6])
vPlusW = v + w
vMinusW = v - w
# compute v+w and v-w

plt.figure(figsize = (6, 6))
a1 = plt.arrow(
    0, 0, v[0], v[1], 
    head_width = 0.3, width = 0.1, color = "k", length_includes_head = True)
a2_1 = plt.arrow(
    0, 0, w[0], w[1], 
    head_width = 0.3, width = 0.1, color = [0.5, 0.5, 0.5], length_includes_head = True)
# v, w (원점에서)
a2_2 = plt.arrow(
    v[0], v[1], w[0], w[1], 
    head_width = 0.3, width = 0.1, color = [0.3, 0.3, 0.3], length_includes_head = True)
# v, w (원점에서)

a3 = plt.arrow(
    0, 0, vPlusW[0], vPlusW[1],
    head_width = 0.3, width = 0.1, color = "steelblue", length_includes_head = True)
# v+w: 원점에서 vPlusW로

a4 = plt.arrow(
    w[0], w[1], vMinusW[0], vMinusW[1],
    head_width = 0.3, width = 0.1, color = "tomato", length_includes_head = True)
# v-w: w 끝점 (4,-6)에서 v 끝점 (1,2)으로
# 평행사변형법 : w + (v-w) = v
# plot all four vectors, using different colors for each

plt.grid(linestyle = "--", linewidth = 0.5)
plt.axis("square")

all_points = np.array([v, w, vPlusW, w + vMinusW])  # 끝점들
lim = max(abs(all_points.flatten())) + 1
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.gca().set_aspect("equal")
# make the axes square, set limits, and add a grid

plt.legend([a1, a2_1, a2_2, a3, a4], ["v", "w(base point)", "w", "v+w", "v-w"])
plt.title("Vectors $v$, $w$ and $v+w$, $v-w$")
plt.show()
```

```python
# 1.2. 차원과 다른 난수 벡터를 사용해 벡터 노름을 계산하는 알고리즘 작성

v = np.array([0, 0, 1])
# test it on a unit-norm vector(단위 벡터)

w = np.array([1, 2, 3])
# non-unit-norm vector, and confirm using np.linalg.norm()

def normOfVector(vector):
    """벡터 노름 계산하는 메서드
    Args:
        vector (np.array): 벡터
    Logics:
        벡터의 각 원소를 제곱한 후, 제곱된 원소들의 합을 구하고, 그 합의 제곱근을 계산하여 벡터 노름을 구함
    Returns:
        float: 벡터 노름"""
    return np.sqrt( np.sum(vector**2) )
    # np.linalg.norm(vector)
```

```python
# 1.2. 차원과 다른 난수 벡터를 사용해 벡터 노름을 계산하는 알고리즘 작성

print(normOfVector(v), np.linalg.norm(v))
print(normOfVector(w), np.linalg.norm(w))
```

```python
# 1.3. 벡터를 입력을 받아 동일한 방향의 단위벡터를 출력하는 함수 구현 

w = np.array([0, 1, 0])
# test on a unit vector

v_1 = np.array([0, 3, 0])
v_2 = np.array([13, -5, 7])
# test on a non-unit vector that is easy to confirm

def createUniVector(vector):
    """벡터의 단위 벡터 계산하는 메서드
    Args:
        vector (np.array): 벡터
    Logics:
        벡터의 각 원소를 벡터 노름으로 나누어 벡터의 단위 벡터를 구함
    Returns:
        np.array: 벡터의 단위 벡터"""
    mu = normOfVector(vector)
    return vector / mu

uw = createUniVector(w)
print(uw)

uv_1 = createUniVector(v_1)
print(uv_1)

uv_2 = createUniVector(v_2)
print(uv_2)
print(np.linalg.norm(v_2), np.linalg.norm(uv_2))
```

```python
# 1.3. 벡터를 입력을 받아 동일한 방향의 단위벡터를 출력하는 함수 구현 

createUniVector(np.zeros( (4, 1) ))
# what happens with the zeros vector?
# 벡터의 노름이 0이므로, 0으로 나누는 상황이 발생하여 오류가 발생함
```

```python
# 1.4. 벡터와 원하는 크기를 입력받고 벡터와 동일한 방향이지만 두 번째 입력에 해당하는 크기를 가진 벡터를 반환하는 함수 구현

w = np.array([1, 0, 0])
# test on a vector that is easy to confirm

def createMagVector(vector, magnitude):
    mu = normOfVector(vector)
    return (magnitude * vector) / mu
```

```python
# 1.4. 벡터와 원하는 크기를 입력받고 벡터와 동일한 방향이지만 두 번째 입력에 해당하는 크기를 가진 벡터를 반환하는 함수 구현

magnitudeVector = createMagVector(vector = w, magnitude = 4)
print(magnitudeVector)
print(np.linalg.norm(w), np.linalg.norm(magnitudeVector))
```

```python
# 1.8. 그림 1-6 

a = np.array([1, 2])
b = np.array([3, 4])
# the vectors

def dotProduct(a, b):
    """벡터의 내적(두 벡터가 얼마나 같은 방향을 보고 있는지)을 계산하는 메서드
    Args:
        a (np.array) : 벡터 a
        b (np.array) : 벡터 b
    Logics:
        벡터 a와 b의 각 원소를 곱한 후, 벡터별로 곱해진 원소들의 합을 구하여 계산
    Returns:
        float: 벡터의 내적
    """
    return sum(x * y for x, y in zip(a, b))
    # np.dot(a, b)

beta = dotProduct(a, b)
print(beta) # (1*3) + (2*4) = 11
```

```python
# 1.8. 그림 1-6 

projvect = b - beta*a
print(projvect)
# 벡터 b에서 벡터 a의 방향으로 beta만큼의 스칼라배로 이동할 투영(projvect) 벡터 계산

print(a.T)

print(dotProduct(a.T, projvect) - beta * dotProduct(a.T, a)) # ???$
```
