````python
# 1.2. 차원과 다른 난수 벡터를 사용해 벡터 노름을 계산하는 알고리즘 작성

v = np.array([0, 0, 1])
# test it on a unit-norm vector(단위 벡터)

w = np.array([1, 2, 3])
# non-unit-norm vector, and confirm using np.linalg.norm()

def normOfVector(vector):
    """벡터 노름 계산하는 메서드
    Args:
        vector (np.array): 벡터
    Logics:
        벡터의 각 원소를 제곱한 후, 제곱된 원소들의 합을 구하고, 그 합의 제곱근을 계산하여 벡터 노름을 구함
    Returns:
        float: 벡터 노름"""
    return np.sqrt( np.sum(vector**2) )
    # np.linalg.norm(vector)
````

````python
# 1.2. 차원과 다른 난수 벡터를 사용해 벡터 노름을 계산하는 알고리즘 작성

print(normOfVector(v), np.linalg.norm(v))
print(normOfVector(w), np.linalg.norm(w))
````

````python
# 1.3. 벡터를 입력을 받아 동일한 방향의 단위벡터를 출력하는 함수 구현 

w = np.array([0, 1, 0])
# test on a unit vector

v_1 = np.array([0, 3, 0])
v_2 = np.array([13, -5, 7])
# test on a non-unit vector that is easy to confirm

def createUniVector(vector):
    """벡터의 단위 벡터 계산하는 메서드
    Args:
        vector (np.array): 벡터
    Logics:
        벡터의 각 원소를 벡터 노름으로 나누어 벡터의 단위 벡터를 구함
    Returns:
        np.array: 벡터의 단위 벡터"""
    mu = normOfVector(vector)
    return vector / mu

uw = createUniVector(w)
print(uw)

uv_1 = createUniVector(v_1)
print(uv_1)

uv_2 = createUniVector(v_2)
print(uv_2)
print(np.linalg.norm(v_2), np.linalg.norm(uv_2))
````

#### 3.3~3.4. 상관계수 직접 구현 및 Numpy.corrcoef(), scipy.stats(), pearsonr()와 성능 비교
* $cor(X, Y) = \rho = \frac{cov(X, Y)}{\sigma_x \cdot \sigma_y}$
* 테스트: 1,000회 이상 for 루프 안에서 500개의 무작위 숫자를 가진 변수 두 개를 생성하고 이들 사이의 상관관계를 계산하고 for루프 시간을 측정

````python
class CorrelationCoefficient:
    """내적을 이용한 상관계수를 구하는 클래스

    정의:
        두 변수 간의 선형적 관계 강도와 방향을 -1에서 1 사이로 나타낸 통계 지표
    원리: 
        공분산(covariance)을 각 변수의 표준편차로 나누어 표준화한 값, 결정계수란 상관계수를 제곱한 값
    유형: 
        피어슨(linear), 스피어만(rank, 정규분포 아닐 때)
    """

    def __init__(self, x, y):
        self.x = x
        self.y = y

       # [to-be] 길이 검증
        if len(self.x) != len(self.y):
            raise ValueError("두 벡터의 길이가 같아야 합니다.")
    
    def rho(self):
        """상관계수 계산 메서드
        numpy.corrcoef(), scipy.stats(), pearsonr()"""
        x_centered = self.x - np.mean(self.x)
        # 데이터 중심화(centering), 각 데이터 벡터에서 평균을 뺀 값
        # 데이터를 평균 0으로 이동시켜 분포를 중심에 맞춤
        y_centered = self.y - np.mean(self.y)

        denominator = self.normofVector(x_centered, y_centered)
        # 중심화된 벡터를 나눌 노름(표준편차)
        numerator = self.dotProduct(x_centered, y_centered)
        # 정규화된 두 벡터 내적

        if denominator == 0:
            # [to-be] 0으로 나누는 경우 미처리(상수 벡터 입력 등)
            raise ValueError("표준편차가 0인 경우 상관계수를 계산할 수 없습니다.")
        else: 
            r = numerator / denominator
            # 상관계수 = 공분산 / 표준편차의 곱 = 내적 / 노름의 곱
            return r

    def dotProduct(self, x: np.array, y: np.array):
        """벡터의 내적(inner product numerator)을 직접 구하는 메서드
        numpy.dot()"""
        projection_result = 0
        # 벡터의 내적을 계산한 결과를 저장할 변수 초기화
        
        for i in range(len(x)):
            projection_result += x[i] * y[i]
            # 이른바 '컬럼 벡터'(1차원 행렬)의 곱
        return projection_result 
    
    def normofVector(self, x: np.array, y: np.array):
        """표준편차의 곱(denominator) 직접 구하는 메서드
        numpy.linalg.norm()"""
        norm_result = 0
        # 벡터의 노름을 계산한 결과를 저장할 변수 초기화

        x_norm = np.sqrt( sum(x_i**2 for x_i in x) )
        # 벡터의 노름
        y_norm = np.sqrt( sum(y_i**2 for y_i in y) )
        norm_result = x_norm * y_norm
        # 벡터의 노름의 곱

        return norm_result
````

````python
# 1. 기본 정확도 테스트

x_01 = np.array([1, 2, 3, 4, 5,], dtype = float)
y_01 = np.array([2, 4, 6, 8, 10], dtype = float)

correlation_object_01 =  CorrelationCoefficient(x_01, y_01)
rho_01 = correlation_object_01.rho()
rho_01_numpy = np.corrcoef(x_01, y_01)[0, 1]

print("Custom rho: ", rho_01)
print("Numpy rho: ", rho_01_numpy)
````

````python
# 2. 음의 상관관계 테스트

y_02 = np.array([10, 8, 6, 4, 2], dtype = float)
rho_02 = CorrelationCoefficient(x_01, y_02).rho()
rho_02_numpy = np.corrcoef(x_01, y_02)[0, 1]

print("Custom rho: ", rho_02)
print("Numpy rho: ", rho_02_numpy)
````

````python
# 3. 랜덤 반복 테스트

random_test_iterations = 10
random_vector_size = 500
rng = np.random.default_rng(seed = 2026)
# 1,000회 이상 for 루프 안에서 500개의 무작위 숫자를 가진 변수 두 개를 생성

for _ in range(random_test_iterations):
    # 이들 사이의 상관관계를 계산하고 for루프 시간을 측정
    x_rand = rng.normal(size = random_vector_size)
    y_rand = rng.normal(size =random_vector_size)
    # 500개의 무작위 숫자로 생성한 두 랜덤 벡터 

    c = CorrelationCoefficient(x_rand, y_rand).rho()
    # 커스텀 클래스로 계산한 랜덤 벡터의 상관계수 rho
    # [error] TypeError: 'numpy.float64' object is not iterable
    n = np.corrcoef(x_rand, y_rand)[0, 1]
    # 넘파이 모듈로 계산한 랜덤 벡터의 상관계수

    assert np.isclose(c, n, atol = 1e-12)
    # 커스텀 클래스와 넘파이 모듈로 계산한 상관계수의 값이 거의 같은지 비교
    print("test passed")
````

#### 3.8~3.9. K-mean Clustering
* k = 3을 사용하여 여러 번 코드 실행 후 최종 클러스터가 결과를 비교 평가
* k = 2, k = 4로 두고 결과 확인 후 비교
* 그룹 중심까지의 거리를 최소화하여 다변량 데이터를 상대적으로 적은 수, k의 범주로 분류하는 비지도학습

> Logic \
> i. 데이터 공간에서 임의의 k개 중심점을 초기화 \
> ii. 중심 = 클래스/범주의 대표점(임의의 차원 수로 일반화) \
> iii. 다음 단계에서 각 데이터 관측치를 각 클래스로 할당 \
> iv. 각 데이터 관측치와 각 중심 사이의 유클리드 거리(L2 distance) 계산 \
> v. 각 데이터 관측치를 가장 가까운 중심의 그룹에 할당 \
> vi. 각 중심을 해당 중심에 할당된 모든 데이터 관측치의 평균으로 갱신 \
> vii. 수렴 기준을 만족할 때까지 4~6단계 반복

> Pseudo code
    ```
    class KMeans
        method __init__(k, max_iterations(최대 반복 횟수), tolerance(허용 오차, 1e-3), seed)
            centers, labels, interia = None

        method initializeCenters(X)
            임의의 k개 점 선택(seed, size k 활용)
        
        method assignClusters(X)
            각 점 - 각 중심 사이의 거리 계산
            가까운 중심으로 할당(labels 활용)

        method updateCenters(X, labels)
            각 클러스터 평균으로 중심 갱신
            빈 클러스터는 임의의 랜덤한 점으로 초기화
        
        method computeInteria(X, labels, centers)
            각 점과 소속 중심 거리 제곱합(SSE) = interia 

        method fit
    ```

#### 3.5. edge 검출기

* 엣지 검출기를 만들어 봅시다. 엣지 검출기의 커널은 [-1, +1]f로 매우 간단합니다. 일정한 시계열 신호 조각(예를 들어, [10, 10])과 커널의 내적은 0입니다. 그러나 신호가 급격히 변화하면 내적이 커집니다.(예를 들어, [1, 10]와의 내적은 9입니다.)
* 여기서 작업할 신호는 평평한 함수입니다. 그래프 A와 B는 각 커널과 신호를 보여줍니다. 이 문제에서 첫 번째 단계는 이 두 시계열을 생성하는 코드를 구현하는 것입니다.
* 다음으로 신호의 시점에 대해 for 루프를 작성합니다. 각 시점에서 커널과 길이가 같은 시계열 데이터 조각과 커널 사이의 내적을 계산합니다. 결과는 그래프 C와 유사해야 합니다. 엣지 검출기는 신호가 평평할 때 0, 신호가 상승할 때 
+1, 신호가 하강할 때 -1을 반환하는 것을 확인해보세요. 

* 예를 들어, 커널을 [0, -1, 1, 0]와 같이 앞 뒤에 0을 추가하면 어떤 영향을 줄까요? → 엣지 검출은 동일하지만, 부드러워짐
* 아니면 커널을 [-1, 1]로 뒤집으면 어떻게 될까요? → 엣지 방향이 반대가 됨
* 커널이 비대칭 [-1, 2]이라면 어떻게 될까요? → 2차 미분, 곡률 변화 탐지

#### 3.6. 시계열을 매끄럽게 만들기 위해 커널 합이 1이 되도록 조정
* 이제 다른 신호와 커널을 사용해서 동일하게 과정을 수행합니다. 목표는 울퉁불퉁한 시계열을 매끄럽게 만드는 것입니다. 시계열을 가우스(정구) 분포에서 생성된 100개의 난수입니다. 커널은 가우스 함수에 근사하는 종 모양의 함수로 숫자 [0, .1, .3, .8, .1, .8, .1, 0]으로 정의되며 커널 합이 1이 되도록 조정되어야  합니다. 그래프 A와 동일해야 하지만 신호는 난수 때문에 그래프 B와 정확히 일치하진 않을 것입니다.
* 이전 연습 문제의 코드를 복사하고 수정해서 시계열을 따라가면서 내적을 계산합니다. 이는 가우스 커널로 필터링된 신호가 됩니다. 여기서 for 루프의 인덱싱에 주의하세요. 그래프 C의 결과의 예제입니다. 필터링된 신호가 원래 신호의 매끄러운 형태라는 것을 알 수 있습니다. 이를 저주파 필터링이라고도 합니다.

#### 3.7. 커널 중앙의 1을 -1로 바꾸고 커널 중앙 평균화, 필터링, 그래프 시각화
* 커널 중앙의 1을 -1로 바꾸고 커널을 중앙 평균화합니다. 그런 다음 다시 필터링하고 그래프를 그립니다. 결과는 더욱 날카로운 신호가 됩니다. 사실 이 커널은 부드러운(낮은 주파수) 특징을 감쇄하고 빠르게 변화하는(높은 주파수) 특징을 강조한 고주파 필터입니다.

````python
# create the kernel (in the book figure I used +1.5)
kernel = np.array([-1,1])

# and the "signal" (a plateau)
signal = np.zeros(30)
signal[10:20] = 1


# plot them
_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kernel,'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-15,15])

axs[1].plot(signal,'ks-')
axs[1].set_title('Time series signal')

# plt.savefig('Figure_03_04ab.png',dpi=300)
plt.show()
````

````python
# initialize the feature map as zeros
featureMap = np.zeros(len(signal))

# loop over the signal and do template-matching (via dot products!)
for t in range(1,len(signal)-1):
  featureMap[t] = np.dot(kernel,signal[t-1:t+1])


# plot the result
_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kernel,'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-15,15])


axs[1].plot(signal,'ks-',label='Signal',linewidth=3)
markers,stemlines,_ = axs[1].stem(range(len(featureMap)),featureMap,
                                  basefmt=' ',linefmt='',markerfmt='o',
                                  label='Edge detection')

plt.setp(stemlines,'color',[.7,.7,.7])
plt.setp(markers,'color',[.7,.7,.7])

axs[1].legend()
# plt.savefig('Figure_03_04c.png',dpi=300)
plt.show()
     
````

````python
# define the kernel (a sorta-kinda Gaussian)
kernel = np.array([0,.1,.3,.8,1,.8,.3,.1,0])
kernel = kernel / np.sum(kernel)

# some handy length parameters
Nkernel = len(kernel)
halfKrn = Nkernel//2


# and the signal
Nsignal = 100
timeseries = np.random.randn(Nsignal)


# plot them
_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kernel,'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-1,Nsignal])

axs[1].plot(timeseries,'ks-')
axs[1].set_title('Time series signal')

# plt.savefig('Figure_03_06ab.png',dpi=300)
plt.show()
````

````python
# make a copy of the signal for filtering
filtsig = timeseries.copy()

# loop over the signal time points
for t in range(halfKrn+1,Nsignal-halfKrn):
  filtsig[t] = np.dot(kernel,timeseries[t-halfKrn-1:t+halfKrn])


# and plot
_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kernel,'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-1,Nsignal])

axs[1].plot(timeseries,color='k',label='Original',linewidth=1)
axs[1].plot(filtsig,'--',color=[.6,.6,.6],label='Smoothed',linewidth=2)
axs[1].legend()

# plt.savefig('Figure_03_06c.png',dpi=300)
plt.show()
````

````python
# define the kernel (a sorta-kinda Gaussian)
kernel = np.array([0,.1,.3,.8,-1,.8,.3,.1,0])
kernel /= np.sum(kernel)
kernel -= np.mean(kernel)

# plot them
_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kernel,'s-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-1,Nsignal])

axs[1].plot(timeseries,'s-')
axs[1].set_title('Time series signal')
plt.show()



# loop over the signal time points
filtsig2 = timeseries.copy()
for t in range(halfKrn+1,Nsignal-halfKrn):
  filtsig2[t] = np.dot(kernel,timeseries[t-halfKrn-1:t+halfKrn])

plt.plot(timeseries,color='k',label='Original',linewidth=1)
plt.plot(filtsig2,color=[.9,.2,.7],label='Sharpened',linewidth=1)
plt.legend()
plt.show()
````
